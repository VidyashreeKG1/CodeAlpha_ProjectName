```python
# Fill null values in the 'checking_balance' column with its mode
# Using the recommended assignment method
df['checking_balance'] = df['checking_balance'].fillna(df['checking_balance'].mode()[0])

# Check for null values after filling
print("Null values after filling:")
print(df.isnull().sum())
```

```python
df.fillna({'months_loan_duration':df['months_loan_duration'].median()}, inplace=True)
df.fillna({'amount':df['amount'].median()}, inplace=True)
df.fillna({'credit_history':df['credit_history'].mode()}, inplace=True)
```

```python
from sklearn.model_selection import train_test_split
import pandas as pd # Import pandas if df is not guaranteed to be defined

# Assuming df is already loaded from a previous cell
# If not, uncomment the line below to load the data
# df = pd.read_csv('credit data.csv')

# Assuming features and target are defined in a previous cell
# Example definitions (replace with your actual features and target):
features = ['months_loan_duration', 'amount', 'percent_of_income', 'age'] # Example using existing numeric columns
target='default' # Assuming 'default' is your target column

x=df[features]
y=df[target]

x_train,x_test,y_train,Y_test=train_test_split(x,y,test_size=0.2,random_state=42)

print("x_train shape:", x_train.shape)
print("x_test shape:", x_test.shape)
```

```python
#Import required classess from sklearn

from sklearn.pipeline import Pipeline

from sklearn.linear_model import LogisticRegression

from sklearn.tree import DecisionTreeClassifier

from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import classification_report
```

```python
models = {
    'LogisticRegression': LogisticRegression(max_iter=1000, random_state=42),
    'DecisionTreeClassifier': DecisionTreeClassifier(random_state=42),
    'RandomForestClassifier': RandomForestClassifier(n_estimators=100, random_state=42)
}
```

```python
from sklearn.metrics import accuracy_score, classification_report

for name, model in models.items():
    print(f"\nModel: {name}")
    pipeline= Pipeline([
           ('calssifier', model)

    ])
    pipeline.fit(x_train, y_train)

    y_pred = pipeline.predict(x_test) # Corrected variable name
    print(f"Accuracy: {accuracy_score(Y_test, y_pred):.4f}") # Corrected print statement
    print("Classification Report:")
    print(classification_report(Y_test, y_pred)) # Corrected print statement
```
